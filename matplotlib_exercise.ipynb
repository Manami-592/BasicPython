## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_data = pd.read_csv('data/hist_data.csv')
plt.hist(df_data['aデータ'], bins=30, color=['orange'], histtype='barstacked', alpha=0.3, label='aデータ')
plt.hist(df_data['bデータ'], bins=30, color=['cyan'], histtype='barstacked', alpha=0.3, label='bデータ')
plt.title('aデータとbデータの温度に関する度数分析')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter = pd.read_csv('data/scatter_data.csv')

plt.scatter(df_scatter[df_scatter['class']==0]['length (cm)'], df_scatter[df_scatter['class']==0]['width (cm)'],color='red', label='class0')
plt.scatter(df_scatter[df_scatter['class']==1]['length (cm)'], df_scatter[df_scatter['class']==1]['width (cm)'],color='blue', label='class1')
plt.scatter(df_scatter[df_scatter['class']==2]['length (cm)'], df_scatter[df_scatter['class']==2]['width (cm)'],color='green', label='class2')

plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.legend()
plt.grid(True)
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line = pd.read_csv('data/line_data.csv')
plt.plot(df_line['期間'], df_line['件数'], marker='o')
plt.grid(True)
plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')
plt.xticks(rotation=30)
```

```python
plt.bar(df_line['期間'], df_line['費用'])
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.title('期間ごとの費用の推移')
plt.xticks(rotation=30)
plt.grid(True)
```
